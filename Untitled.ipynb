```python
def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    
    cut_off = IQR * 1.5
    
    outliers_mask = (series < (Q1 - cut_off)) | (series > (Q3 + cut_off))
    return outliers_mask.sum()
```

```python
count_outliers(df["salary"])
```

```python
# 1. Calcula os limites
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

filtro_outliers = (df['salary'] < limite_inferior) | (df['salary'] > limite_superior)

df[filtro_outliers]
```

```python
# Identifiquei 4 dados outliers que podem indicar erros na conversão da moeda para Dolar
df = df.drop(df[filtro_outliers].index)
```
